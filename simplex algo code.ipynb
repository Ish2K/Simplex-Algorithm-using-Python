```python
number_of_variables = int(input("Number of Variables : "))
c = [0]*(number_of_variables) 
for i in range(number_of_variables):
    c[i] = int(input("Coefficient of Variable x_"+str(i+1)+": "))

number_of_constraints = int(input("Number of Constraints: "))

ZjCj = []
A = []

for i in range(number_of_constraints):
    c.append(0) #slack values
    row = []
    print("Constraint "+ str(i+1) +" Values: ")
    for j in range(number_of_variables):
        val = int(input("Coefficient of Variable x_"+str(j+1)+": "))
        row.append(val)
    for j in range(number_of_constraints): #as number of slack variables=number of constraints
        if(i==j):
            row.append(1)
        else:
            row.append(0)
    bound = int(input("Upper Bound : "))
    row.append(bound)
    A.append(row)

#print(A)

for i in c:
    ZjCj.append(-i)

ZjCj.append(0) #solution
itr = 1

unbounded = False
optimal = False

var_name = ["s_"+str(i) for i in range(1,number_of_constraints+1)]


while((not unbounded) and (not optimal)):
    print("Iteration :",itr)
    
    optimal = True
    mini =  10**9
    idx = 0
    pivot_clm = -1
    for j in ZjCj:
        if(j<0 and mini>j):
            optimal = False
            mini = j
            pivot_clm = idx
        idx+=1
    if(optimal):
        print("Value of Maximum Objective Function =",ZjCj[-1])
        curr_idx = 0
        for m in A:
            print(var_name[curr_idx]+" = "+str(m[-1]))
            curr_idx+=1
        break
    
    ratios = []
    mini = 10**9
    pivot_row = -1
    curr = 0
    for m in A:
        if(m[pivot_clm]==0):
            curr+=1
            continue
        tmp = m[-1]/m[pivot_clm]
        if(tmp>=0):
            ratios.append(tmp)
            if(tmp<mini):
                mini = tmp
                pivot_row = curr
        curr+=1

    if(len(ratios)==0):
        print("Solution is Unbounded!")
        break
    
    key_elem = A[pivot_row][pivot_clm]
    
    var_name[pivot_row] = "X_"+str(pivot_clm+1)
    for k in range(len(A[pivot_row])):
        
        A[pivot_row][k] /= key_elem
    
    for a in range(len(A)):
        if(a==pivot_row):
            continue
        factor = A[a][pivot_clm]
        for b in range(len(A[a])):
            A[a][b] -= factor*A[pivot_row][b]
    
    factor = ZjCj[pivot_clm]
    for a in range(len(ZjCj)):
        ZjCj[a] -= factor*A[pivot_row][a]
    
    for i in range(len(ZjCj)):
        print(ZjCj[i],end = " ")
    print()
    for i in range(len(A)):
        for j in range(len(A[i])):
            print(A[i][j],end=" ")
        print()
    itr+=1
```
